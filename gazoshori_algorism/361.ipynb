```python
import cv2
import numpy as np

def main():
    
    t = 127
    
    img = cv2.imread("./sample/3-6/input.jpg")
    
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    th1 = gray.copy()
    
    th1[gray < t] = 0
    th1[gray >= t] = 255
    
    #表示
    im4p = cv2.cvtColor(th1, cv2.COLOR_BGR2RGB)
    plt.imshow(im4p)
    
main()
```

```python
import cv2
import numpy as np

def threshold(src, ksize = 3, c=2):
    #畳み込み演算をしない領域の幅
    d = int((ksize-1)/2)
    h, w = src.shape[0], src.shape[1]
    
    # 出力画像用の配列
    dst =  np.empty((h,w))
    dst.fill(255)
    
    n = ksize**2
    
    for y in range (0,h) :
        for x in range(0, w):
            t= np.sum(src[y-d:y+d+1, x-d:x+d+1])/n
            
            if (src[y][x] < t - c): 
                dst[y][x] = 0
            else:
                dst[y][x] = 255
    
    return (dst)


def main():
    img = cv2.imread("./sample/3-6/input.jpg")
    
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    dst1 = threshold(gray, ksize=11, c=2)

    dst2 = dst1.astype(np.uint8)
    im4p = cv2.cvtColor(dst2, cv2.COLOR_BGR2RGB)
    plt.imshow(im4p)
#     cv2.imwrite("output1_yoshi.jpg",dst1)

main()
```
