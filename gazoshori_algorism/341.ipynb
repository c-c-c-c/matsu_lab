```python
# p115
import cv2
import numpy as np
from matplotlib import pyplot as plt

def main():
    img = cv2.imread("./sample/3-4/input.jpg")
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    hist, bins = np.histogram(gray.ravel(), 256 , [0,256])
    print(hist)

    plt.xlim(0,255)
    plt.plot(hist)
    plt.grid()
    plt.show()
    

main()
```

```python
import cv2
import numpy as np

def main():
    img = cv2.imread("./sample/3-4/input.jpg")

    # 
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # 
    gamma = 2.0
    
    imax = gray.max()
    
    # 
    gray = imax * (gray/imax)**(1/gamma)
    
    # 
    gray_int = gray.astype(np.uint8)
    im4p = cv2.cvtColor( gray_int, cv2.COLOR_BGR2RGB )
    plt.imshow(im4p)
   
main()
```
